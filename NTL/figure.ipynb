```python
import matplotlib.pyplot as plt
import pandas as pd 
import matplotlib.dates as mdates
from matplotlib.font_manager import FontProperties
```

```python
df = pd.read_csv('Pro_ratio_DNB_BRDF_Corrected_NTL.csv')
df['Month'] = pd.to_datetime(df['Month'] )
df = df[df['Month'].dt.year>2016]
df = df.set_index('Month')
```

```python
font = FontProperties(family='sans-serif', size=5)
font2 = FontProperties(family='sans-serif', size=5)
```

```python
# 创建图形和子图
fig = plt.figure(figsize=(3.5,3), dpi=300) # 设置图形尺寸为180 mm x 90 mm，分辨率为300 dpi
fig.patch.set_facecolor('white')  # 设置背景颜色为白色

# 绘制第一个位置的图形
ax1 = fig.add_subplot(2, 2, 1)
d = df[df['province']=='上海']
# 绘制线图
ax1.plot(pd.to_datetime(d.index), d['C_BRDF_R'], color='#607B8B', linewidth=0.6, label='Commercial')
ax1.plot(pd.to_datetime(d.index), d['R_BRDF_R'], color='#9FB6CD', linewidth=0.6, label='Residential', linestyle='-.')
ax1.plot(pd.to_datetime(d.index), d['T_BRDF_R'], color='#339966', linewidth=0.6, label='Active', linestyle='--')
ax1.legend(prop=font, title=None)
# 设置X轴刻度定位器
ax1.xaxis.set_major_locator(mdates.MonthLocator(bymonth=1, bymonthday=1))
# 设置X轴日期格式
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
# 设置X轴范围
ax1.set_xlim(pd.to_datetime("2017-01-01"), pd.to_datetime("2022-12-31") )
# 设置X轴刻度定位器
ax1.xaxis.set_major_locator(mdates.YearLocator())
# 设置Y轴范围和刻度
ax1.set_ylim(0, 200)
# 设置图形样式和标签
ax1.set_ylabel('')
ax1.set_title('Shanghai NTL', loc='center', y=0.95, fontproperties=font2)

# 设置 X 轴刻度标签字体
for xtick in ax1.get_xticklabels():
    xtick.set_fontproperties(font)
# 设置 Z 轴刻度标签字体
for ytick in ax1.get_yticklabels():
    ytick.set_fontproperties(font)
ax1.set_position([0.1, 0.55, 0.4, 0.4]) 

ax2 = fig.add_subplot(2, 2, 2)
d = df[df['province']=='北京']
# 绘制线图
ax2.plot(pd.to_datetime(d.index), d['C_BRDF_R'], color='#607B8B', linewidth=0.6, label='Commercial')
ax2.plot(pd.to_datetime(d.index), d['R_BRDF_R'], color='#9FB6CD', linewidth=0.6, label='Residential', linestyle='-.')
ax2.plot(pd.to_datetime(d.index), d['T_BRDF_R'], color='#339966', linewidth=0.6, label='Active', linestyle='--')

# 设置X轴刻度定位器
ax2.xaxis.set_major_locator(mdates.MonthLocator(bymonth=1, bymonthday=1))
# 设置X轴日期格式
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
# 设置X轴范围
ax2.set_xlim(pd.to_datetime("2017-01-01"), pd.to_datetime("2022-12-31") )
# 设置X轴刻度定位器
ax2.xaxis.set_major_locator(mdates.YearLocator())
# 设置Y轴范围和刻度
ax2.set_ylim(0, 200)
# 设置图形样式和标签
ax2.set_ylabel('')
ax2.set_title('Beijing NTL', loc='center', y=0.95, fontproperties=font2)

# 设置 X 轴刻度标签字体
for xtick in ax2.get_xticklabels():
    xtick.set_fontproperties(font)
# 设置 Z 轴刻度标签字体
for ytick in ax2.get_yticklabels():
    ytick.set_fontproperties(font)
ax2.set_position([0.58, 0.55, 0.4, 0.4])# [left, bottom, width, height]

ax3 = fig.add_subplot(2, 2, 3)
d = df[df['province']=='天津']
# 绘制线图
ax3.plot(pd.to_datetime(d.index), d['C_BRDF_R'], color='#607B8B', linewidth=0.6, label='Commercial')
ax3.plot(pd.to_datetime(d.index), d['R_BRDF_R'], color='#9FB6CD', linewidth=0.6, label='Residential', linestyle='-.')
ax3.plot(pd.to_datetime(d.index), d['T_BRDF_R'], color='#339966', linewidth=0.6, label='Active', linestyle='--')

# 设置X轴刻度定位器
ax3.xaxis.set_major_locator(mdates.MonthLocator(bymonth=1, bymonthday=1))
# 设置X轴日期格式
ax3.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
# 设置X轴范围
ax3.set_xlim(pd.to_datetime("2017-01-01"), pd.to_datetime("2022-12-31") )
# 设置X轴刻度定位器
ax3.xaxis.set_major_locator(mdates.YearLocator())
# 设置Y轴范围和刻度
ax3.set_ylim(0, 200)
# 设置图形样式和标签
ax3.set_ylabel('')
ax3.set_title('Tianjin NTL', loc='center', y=0.95, fontproperties=font2)
# 设置 X 轴刻度标签字体
for xtick in ax3.get_xticklabels():
    xtick.set_fontproperties(font)
# 设置 Z 轴刻度标签字体
for ytick in ax3.get_yticklabels():
    ytick.set_fontproperties(font)
ax3.set_position([0.1, 0.06, 0.4, 0.4])# [left, bottom, width, height]

ax4 = fig.add_subplot(2, 2, 4)
d = df[df['province']=='重庆']
# 绘制线图
ax4.plot(pd.to_datetime(d.index), d['C_BRDF_R'], color='#607B8B', linewidth=0.6, label='Commercial')
ax4.plot(pd.to_datetime(d.index), d['R_BRDF_R'], color='#9FB6CD', linewidth=0.6, label='Residential', linestyle='-.')
ax4.plot(pd.to_datetime(d.index), d['T_BRDF_R'], color='#339966', linewidth=0.6, label='Active', linestyle='--')

# 设置X轴刻度定位器
ax4.xaxis.set_major_locator(mdates.MonthLocator(bymonth=1, bymonthday=1))
# 设置X轴日期格式
ax4.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
# 设置X轴范围
ax4.set_xlim(pd.to_datetime("2017-01-01"), pd.to_datetime("2022-12-31") )
# 设置X轴刻度定位器
ax4.xaxis.set_major_locator(mdates.YearLocator())
# 设置Y轴范围和刻度
ax4.set_ylim(0, 200)
# 设置图形样式和标签
ax4.set_ylabel('')
ax4.set_title('Chongqing NTL', loc='center', y=0.95, fontproperties=font2)

# 设置 X 轴刻度标签字体
for xtick in ax4.get_xticklabels():
    xtick.set_fontproperties(font)
# 设置 Z 轴刻度标签字体
for ytick in ax4.get_yticklabels():
    ytick.set_fontproperties(font)
ax4.set_position([0.58, 0.06, 0.4, 0.4]) # [left, bottom, width, height]


plt.savefig('NTL_pro1.tiff', dpi=300, format="tiff")
# 显示图形
#plt.show()
```
